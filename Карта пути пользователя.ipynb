# Карта пути пользователя: Система оценки изменений без A/B-тестов

## Роли пользователей

### ПРОДУКТОВЫЙ АНАЛИТИК
**Цель:** Быстро и достоверно оценить эффект от изменения без запуска полноценного A/B-теста  
**Мысли:** "Сработала ли наша гипотеза?", "Можно ли доверять этой оценке?"  
**Барьеры:** Сложность методов причинно-следственного анализа, необходимость доказывать достоверность результата продукт-менеджерам

### ПРОДУКТ-МЕНЕДЖЕР
**Цель:** Получить быстрое подтверждение гипотезы для принятия решения о дальнейшей разработке  
**Мысли:** "Стоит ли нам тратить время на эту фичу?", "Насколько велик ожидаемый эффект?"  
**Барьеры:** Недоверие к методам, отличным от A/B-теста; сложность интерпретации статистических выводов

### DATA ENGINEER
**Цель:** Обеспечить аналитика качественными и доступными данными для анализа  
**Мысли:** "Какие именно данные и за какой период нужны?", "Достаточно ли производителен пайплайн формирования дампов?"  
**Барьеры:** Высокая нагрузка от неоптимизированных запросов, разнородность источников данных

## Ключевые метрики успеха

- **Время анализа:** Сокращение с 1-2 недель (A/B-тест) до 1-2 дней
- **Точность:** Ошибка оценки менее 5% по сравнению с результатами A/B-тестов

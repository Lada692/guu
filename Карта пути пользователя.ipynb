# Карта пути пользователя: Система оценки изменений без A/B-тестов

## Роли пользователей

### ПРОДУКТОВЫЙ АНАЛИТИК
**Цели по SMART:**
- **Конкретная:** Оценить эффект изменения интерфейса меню для продавцов с точностью ±3%
- **Измеримая:** Получить статистически значимый результат (p-value < 0.05) с доверительным интервалом
- **Достижимая:** Использовать готовый Jupyter Notebook с документацией
- **Релевантная:** Сократить время проверки гипотезы с 14 до 2 дней
- **Ограниченная по времени:** Провести оценку в течение 2 рабочих дней

### ПРОДУКТ-МЕНЕДЖЕР
**Цели по SMART:**
- **Конкретная:** Принять решение о дальнейшей разработке фичи на основе оценки эффекта
- **Измеримая:** Получить отчет с визуализацией и четкой рекомендацией (запускать/не запускать A/B-тест)
- **Достижимая:** На основе данных от аналитика за 2 дня
- **Релевантная:** Снизить риски разработки неэффективных фич на 40%
- **Ограниченная по времени:** Принять решение в течение 1 дня после получения отчета

### DATA ENGINEER
**Цели по SMART:**
- **Конкретная:** Обеспечить доступ к данным трекера за последние 90 дней
- **Измеримая:** Предоставить дампы данных объемом до 50 ГБ в формате Parquet
- **Достижимая:** Используя существующие ETL-пайплайны
- **Релевантная:** Обеспечить воспроизводимость анализа
- **Ограниченная по времени:** В течение 4 часов с момента запроса

## Ключевые метрики успеха (SMART)

### Для продуктового аналитика
- **Сокращение времени анализа:** с 14 дней до 2 дней (±1 день)
- **Точность оценки:** ошибка не более ±3% при валидации на исторических A/B-тестах
- **Статистическая значимость:** достижение p-value < 0.05 в 90% случаев при наличии реального эффекта

### Для продукт-менеджера
- **Скорость принятия решений:** получение обоснованной рекомендации в течение 3 дней
- **Качество решений:** снижение количества запущенных неэффективных фич на 40%
- **Прозрачность:** 100% отчетов содержат визуализацию и интерпретацию результатов

## Бизнес-показатели эффективности

1. **Экономия времени:** 12 дней на гипотезу × 10 гипотез в квартал = 120 человеко-дней
2. **Снижение рисков:** предотвращение разработки 4 неэффективных фич в квартал
3. **Ускорение итераций:** увеличение скорости проверки гипотез на 85%
4. **Рост удовлетворенности:** достижение 85% NPS среди пользователей системы
